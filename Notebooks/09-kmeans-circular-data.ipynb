# K-Means - Circular Data

```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("circular_data.csv")
df.sample(10)
```

```python
fig = px.scatter(df,
                    x="x",
                    y="y",
                    color="cluster",
                    height=500)
fig.show()
```

## K-Means

```python
X = df[['x', 'y']]

kmeans = KMeans(n_clusters=3, random_state=0, n_init='auto')
kmeans.fit(X)
```

```python
df["predicted"] = kmeans.predict(X)
df.sample(10)
```

## Visualization

```python
fig = px.scatter(df,
                    x="x",
                    y="y",
                    color="predicted",
                    height=500)
fig.show()
```
