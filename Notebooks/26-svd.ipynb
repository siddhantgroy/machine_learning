# Singular Value Decomposition (SVD)

```python
import numpy as np
```

```python
X = np.random.normal(size=[20,18])
X.shape
```

```python
X
```

```python
U, S, Vt = np.linalg.svd(X, full_matrices=False)
print(U.shape, S.shape, Vt.shape)
```

```python
# recreating the X matrix by multiplying U, S, and Vt

# X_prime = np.matmul(U, np.matmul(np.diag(S), Vt))
X_prime = np.dot(U, np.dot(np.diag(S), Vt))
X_prime.shape
```

```python
# comparing the original matrix with the recreated matrix
# np.allclose compare two matrices element by element

np.allclose(X, X_prime)
```
