# Support Vector Machine (SVM) Classifier - Breast Cancer

```python
from sklearn import datasets
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_breast_cancer.html

cancer = datasets.load_breast_cancer()
print(f"Features: {cancer.feature_names}")
print("")
print(f"Target: {cancer.target_names}")
```

```python
df = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)
df.head()
```

```python
df.shape
```

```python
# 0: malignant (bad), 1: benign (Not bad)

df["target"] = cancer.target
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

## Data Split

```python
X = df.drop(columns=["target"])
y = df["target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Train: {X_train.shape}")
print(f"Test: {X_test.shape}")
```

## Modeling / Training

```python
model = SVC(kernel="linear") # linear, rbf, sigmoid, poly
model.fit(X_train, y_train)
```

## Evaluation

```python
y_predicted = model.predict(X_test)
```

### Accuracy

```python
print(f"Accuracy: {metrics.accuracy_score(y_test, y_predicted)}")
```

### Confusion Matrix

```python
cm = metrics.confusion_matrix(y_test, y_predicted, labels=model.classes_)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()
plt.show()
```
