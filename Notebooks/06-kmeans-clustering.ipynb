# K-Means Clustering

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=30, centers=3, n_features=2,
                  random_state=0)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=y, marker="o")
plt.show()
```

## K-Means

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)
```

```python
centroids = kmeans.cluster_centers_
centroids
```

```python
kmeans.labels_
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_, marker="o", s=200, alpha=0.5, label="Predicted")
plt.scatter(X[:,0],X[:,1],c=y, marker="*", alpha=0.5, label="Actual")
plt.scatter(centroids[:,0],centroids[:,1],c="red", marker="+", label="Centroid")
plt.legend()
plt.show()
```
